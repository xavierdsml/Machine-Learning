Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Creation of Dataset - quadratic equation y = 0.5X^2 + 1.5X + 2 + outliers

```python
#value are already in scalling means small 0-1
X = 6 * np.random.rand(100,1)-3
y = 0.5* X**2 + 1.5*X + 2 + np.random.randn(100,1)
```

Show the DataSet

```python
plt.scatter(X,y, color = 'g')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 11)
```

```python
from sklearn.linear_model import LinearRegression
linearRegression = LinearRegression()
```

```python
linearRegression.fit(X_train, y_train)
```

```python
y_predict = linearRegression.predict(X_test)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_test, y_predict)
plt.show()
```

```python
#check the preformance of the model import r2 
from sklearn.metrics import r2_score
score = r2_score(y_test, y_predict)
score
```

How to do with Polynomial Regression beacuse this data is not in Linear Realtionship this is in Non-Linear Realtionship 

```python
#POLYNOMIAL REGRESSION
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2, include_bias = True)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
#now apply linear regression

linearRegression.fit(X_train_poly, y_train)
```

```python
y_predict_poly = linearRegression.predict(X_test_poly)
```

```python
X_range = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1, 1)

X_range_poly = poly.transform(X_range)
y_range_pred = linearRegression.predict(X_range_poly)

plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_range, y_range_pred, color='red', label='Polynomial Fit')

plt.legend()
plt.show()
```

```python
##check the accuracy of the model 

from sklearn.metrics import r2_score

score = r2_score(y_test, y_predict_poly)
score
```
